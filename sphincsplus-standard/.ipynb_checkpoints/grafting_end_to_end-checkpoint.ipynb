```python
def convert_hex_string(hex_string):
    # Ensure the input length is even
    if len(hex_string) % 2 != 0:
        raise ValueError("Hex string length must be even.")

    # Split into pairs and format
    formatted_hex = ", ".join(f"0x{hex_string[i:i+2]}" for i in range(0, len(hex_string), 2))
    
    return formatted_hex

# Example usage
hex_input = "B8870187C7E4301977538435A4EC73EDC24F66469AF70BDA62247D6CE6E20D4D45215997711A717A4CC2061E206CF7C254FC8721E6E8A4B35F5C49CF766E38C2964302B9BED74C8DF1800924F4F168FFC0C71BFB8F4A2D3E1715AF5E9F0336B86567EDC8FFD72143922D33DB909688D6431CE3E8D11CAE27AF6300E0A044D6CA855346DB1834E4E97A470B94EE6C4532924F2BDED95783C90724B21F22BCAB1BE30F5672F58E80C19491203B04A360F723FF22B64001C672E18E15ECBFBEEB60BEEC7814F27A9A0CBFBF5A8B3A10F43C1B825A7A313D9EBB0FACF7ABE6D9C2FE0EB36E98D73030C719C73341F5D02FC3FBBDAA377C88441972959C107F3F2798A8BB581D33C867FB6B668ADD7C2F3DBE4E1BCB867830E845CABF6A32CAAC7093387AE502D3331F0575DAB00D3F45DA7DB15AB6E019CC99119E409AE917DBD2476D30C4A51AB759EEF15BEACD69C0748BBB8D9A58E4A698BEDBEF204A4C34F361DE3D9B63FD02F1723ACE9877846AC2958467BD22BC26AE75A66F76CFFF7A9CD8C6D7D0D7E3857A4ABDF5353C3957E10928161B821A9D85B933D313FA3D8412C7027F53661C5F45D5048D943A44BBEA56B6DD9793A5F344D643A47C86FDD737CD42FA62A2B9CF9D39778554430FC4CF77B4531652D09E8DC5CDB04B1F2D977A27EB7E3F78ED02F60CADC0ED32EA320C9233A98DB9DCC15C54EAAE1E1EB45A4E3B3480764B2DD6FCB6AB17E6E919BE3841921A09864974A666656430EBE9897280804C7240FB1324E84A0AF129C555FF27"
output = convert_hex_string(hex_input)
print(output)  # Output: 0xAB, 0xB8, 0xAC, 0x57
```
